```python
# Import the relavant libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the World Population Data
dataset = pd.read_csv('world_pop.csv')
```

```python
# Display the first five columns in the dataset

dataset.head()
```

```python
# To clean up the data a little, I want to drop the first column and rename the other columns

# Drop the first column (index 0)
dataset_edited = dataset.drop(dataset.columns[0], axis=1)
```

```python
# Rename the columns
dataset_edited = dataset_edited.rename(columns=lambda x: x.replace('year_', ''))
```

```python
dataset_edited.head()
```

```python
dataset_edited.dtypes
```

```python
dataset_edited.tail()
```

```python
# Calculate the sum of population for each year
year_sums = dataset_edited.iloc[:, 1:].sum(axis=0)

# Create a new row to store the total population
total_row = pd.DataFrame(year_sums).T
total_row.insert(0, 'country', 'Total')
total_row.reset_index(drop=True, inplace=True)

# Append the total row to the dataset
dataset_edited = dataset_edited.append(total_row, ignore_index=True)
```

```python
dataset_edited.tail()
```

```python
# Set the option to display numbers with commas
pd.options.display.float_format = '{:,.0f}'.format
```

```python
dataset_edited.tail()
```

```python
# Visualizing population growth

# Extract relevant columns
years = dataset_edited.columns[1:]
population = dataset_edited.iloc[-1, 1:]

# Create a line plot using seaborn
plt.figure(figsize=(10, 5))
sns.lineplot(x=years, y=population, color='blue')
plt.xlabel('Year')
plt.ylabel('Population (billion)')
plt.xticks(rotation=90)
plt.grid(False)


# Save the plot as a JPEG file
plt.savefig('population_growth.jpg', format='jpeg')


plt.show()
```

```python
# Define the column names for the years in your dataset
year_columns = ['1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970', 
                '1971', '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979', '1980', '1981', 
                '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', 
                '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', 
                '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', 
                '2015', '2016', '2017', '2018', '2019', '2020']

# Group the data by continent and sum the population figures for each year
continent_pop = dataset_edited.groupby('Continent')[year_columns].sum()

# Reset the index to make the continent column a regular column
continent_pop = continent_pop.reset_index()

# Print the resulting dataset
print(continent_pop)
```

```python
# Create a list of continents
continents = ['Africa', 'Asia', 'Europe', 'North America', 'Oceania', 'South America']

# Plot the population trend for each continent
plt.figure(figsize=(10, 5))
for continent in continents:
    # Get the population data for the continent
    continent_data = continent_pop[continent_pop['Continent'] == continent].drop('Continent', axis=1)
    
    # Reshape the data to long format
    continent_data = continent_data.melt(var_name='Year', value_name='Population')
    
    # Plot the population data using seaborn lineplot
    sns.lineplot(data=continent_data, x='Year', y='Population', label=continent)

# Set the x-axis and y-axis labels
plt.xlabel('Year')
plt.ylabel('Population (billion)')
plt.xticks(rotation=90)

# Set the title of the plot
plt.legend()

# Save the plot as a JPEG file
plt.savefig('continental_growth.jpg', format='jpeg')

plt.show()
```

```python
# Set the 'Continent' column as the index
continent_pop.set_index('Continent', inplace=True)

# Calculate the yearly population changes and remove the first column (1960) from the DataFrame 
continent_change = continent_pop.diff(axis=1)
continent_change = continent_change.drop(columns='1960')

# Reset the index
continent_change.reset_index(inplace=True)
```

```python
continent_change
```

```python
# Create a list of continents
continents = ['Africa', 'Asia', 'Europe', 'North America', 'Oceania', 'South America']

# Plot the population trend for each continent
plt.figure(figsize=(10, 5))
for continent in continents:
    # Get the population data for the continent
    continent_change_data = continent_change[continent_change['Continent'] == continent].drop('Continent', axis=1)
    
    # Reshape the data to long format
    continent_change_data = continent_change_data.melt(var_name='Year', value_name='Population')
    
    # Plot the population data using seaborn lineplot
    sns.lineplot(data=continent_change_data, x='Year', y='Population', label=continent)

# Set the x-axis and y-axis labels
plt.xlabel('Year')
plt.ylabel('Population (million)')
plt.xticks(rotation=90)

# Save the plot as a JPEG file
plt.savefig('continental_change.jpg', format='jpeg')

plt.show()
```
