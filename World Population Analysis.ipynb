```python
# Import the relavant libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the World Population Data
dataset = pd.read_csv('world_pop.csv')
```

```python
# Display the first five columns in the dataset

dataset.head()
```

```python
# To clean up the data a little, I want to drop the first column and rename the other columns

# Drop the first column (index 0)
dataset_edited = dataset.drop(dataset.columns[0], axis=1)
```

```python
# Rename the columns
dataset_edited = dataset_edited.rename(columns=lambda x: x.replace('year_', ''))
```

```python
# Calculate the sum of population for each year
year_sums = dataset_edited.iloc[:, 1:].sum(axis=0)

# Create a new row to store the total population
total_row = pd.DataFrame(year_sums).T
total_row.insert(0, 'country', 'Total')
total_row.reset_index(drop=True, inplace=True)

# Append the total row to the dataset
dataset_edited = dataset_edited.append(total_row, ignore_index=True)
```

```python
# Convert the years to integers
continent_change = continent_change.astype({'Continent': str})
continent_change.iloc[:, 1:] = continent_change.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

# Create a new DataFrame to store the decade sums
decade_sums = pd.DataFrame({'Continent': continent_change['Continent']})

# Create six new columns for each decade's sum
for i in range(1, 7):
    column_name = f'Decade_{i}_Sum'
    year_range = f"{1960 + (i - 1) * 10}-{1969 + (i - 1) * 10}"
    decade_sum = continent_change.iloc[:, (i - 1) * 10 + 1: i * 10 + 1].sum(axis=1)
    decade_sums[year_range] = decade_sum
```

```python
# Transpose the DataFrame
decade_sums_transposed = decade_sums.set_index('Continent').T

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Width of each bar
bar_width = 0.15

# Position of each bar on the x-axis
x_pos = list(range(len(decade_sums_transposed)))

# List of continents
continents = decade_sums_transposed.columns

# Plotting the grouped bars
for i, continent in enumerate(continents):
    x_positions = [pos + i * bar_width for pos in x_pos]
    ax.bar(x_positions, decade_sums_transposed[continent], width=bar_width, label=continent)

# Adjusting x-axis and ticks
ax.set_xticks([pos + 2 * bar_width for pos in x_pos])
ax.set_xticklabels(decade_sums_transposed.index)
ax.set_xlabel('Decades')
ax.set_ylabel('Population')

# Adding a legend
ax.legend()

# Save the plot as a JPEG file
plt.savefig('decade_change_graph.jpg', format='jpeg')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Checking out the visualization in line graph
# Set up the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the lines for each continent
for continent in decade_sums_transposed.columns:
    ax.plot(decade_sums_transposed.index, decade_sums_transposed[continent], label=continent)

# Adjusting x-axis and ticks
ax.set_xticks(decade_sums_transposed.index)
ax.set_xlabel('Decades')
ax.set_ylabel('Population')

# Adding a legend
ax.legend()

# Save the plot as a JPEG file
plt.savefig('continental_change2.jpg', format='jpeg')

# Show the plot
plt.tight_layout()
plt.show()
```
